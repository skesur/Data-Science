```python
import plotly.express as px
import pandas as pd
```

```python
df = px.data.tips()

print(df.head())
```

```python
fig = px.line(
    df,
    x="total_bill",
    y="tip",
    title="Line Chart",
    color="sex"   # grouping
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="total_bill",
    y="tip",
    color="sex",
    size="size",
    hover_data=["time"],
    symbol="sex"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="total_bill",
    y="tip",
    size="size",        # dynamic size
    color="total_bill"  # dynamic color
)

fig.show()
```

```python
color_map = {
    "Male": "blue",
    "Female": "pink"
}

fig = px.scatter(
    df,
    x="total_bill",
    y="tip",
    color="sex",
    color_discrete_map=color_map
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="total_bill",
    y="tip",
    color="total_bill",
    color_continuous_scale="viridis"
)

fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="total_bill",
    y="tip",
    z="size",
    size="size",        # dynamic size
    color="total_bill"  # dynamic color
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="total_bill",
    y="tip",
    size="size",
    size_max=40,
    opacity=0.6
)

fig.show()
```

```python
fig = px.bar(
    df,
    x="day",
    y="total_bill",
    color="sex",
    barmode="group"   # or "stack"
)

fig.show()
```

```python
fig = px.bar(
    df,
    x="day",
    y="total_bill",
    color="sex",
    barmode="stack"   # or "stack"
)

fig.show()
```

```python
fig = px.bar(
    df,
    x="day",
    y="total_bill",
    color="sex",
    barmode="relative"   # or "stack"
)

fig.show()
```

```python
fig = px.bar(
    df,
    x="day",
    y="total_bill",
    color="day",
    color_discrete_sequence=["red", "blue", "green", "orange"],
    hover_data=["sex", "day"]
)

fig.show()
```

```python
fig = px.histogram(
    df,
    x="total_bill",
    nbins=5,
    color="sex"
)

fig.show()
```

```python
fig = px.box(
    df,
    x="day",
    y="total_bill",
    color="sex",
    points="all"   # show all points
)

fig.show()
```

```python
fig = px.pie(
    df,
    names="day",
    values="total_bill",
    title="Pie Chart"
)

fig.show()
```

```python
corr = df.corr(numeric_only=True)

fig = px.imshow(
    corr,
    text_auto=True,
    color_continuous_scale="RdBu"
)

fig.show()
```

```python
fig = px.area(
    df,
    x="total_bill",
    y="tip",
    color="sex"
)

fig.show()
```

```python
fig = px.scatter_matrix(
    df,
    dimensions=["total_bill", "tip", "size"],
    color="sex"
)

fig.show()
```
