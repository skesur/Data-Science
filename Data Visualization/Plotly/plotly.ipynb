```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default="notebook"
```

```python
df = pd.DataFrame({
    "De":["IT","IT","HR","HR","Finance","Finance"],
    "Gender":["M","F","M","F","M","F"],
    "sal":[50000,-50000,45000,48000,70000,72000]
})
df
```

```python
fig = px.bar(df,x="De",color="Gender",y="sal")
fig.show()
```

```python
fig = px.bar(df,barmode="stack",color="Gender",x="De",y="sal")
fig.show()
```

```python
fig = px.bar(df,barmode="relative",color="Gender",x="De",y="sal")
fig.show()
```

```python
fig = px.bar(df,barmode="group",color="Gender",x="De",y="sal")
fig.show()
```
