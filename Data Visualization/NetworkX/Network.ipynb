```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_node("A")
G.add_node("B")

G.add_edge("A", "B")

G.add_nodes_from(["C", "D", "E"])

G.add_edges_from([
    ("A", "C"),
    ("B", "D"),
    ("C", "D"),
    ("C", "E")
])

nx.draw_networkx(G)
plt.show()
```

```python
print("Nodes:", G.nodes())
print("Edges:", G.edges())
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print(G.degree())
print(dict(G.degree()))
```

```python
nx.draw(
    G,
    with_labels=True,
    node_color="lightblue",
    node_size=2000,
    font_size=12,
    edge_color="gray"
)

plt.show()
```

```python
nx.has_path(G, "A", "D")
```

```python
nx.shortest_path(G, "A", "D")
```

```python
nx.shortest_path_length(G, "A", "D")
```

```python
G = nx.Graph()

G.add_edge("A","B",weight=1)
G.add_edge("B","C",weight=2)
G.add_edge("A","C",weight=5)

nx.shortest_path(G, "A", "C", weight="weight")
```

```python
print(list(G.neighbors("A")))
```

```python
G = nx.Graph()

G.add_edges_from([("A","B"),("C","D")])

components = list(nx.connected_components(G))

print(components)
```

```python
import pandas as pd

df = pd.DataFrame({
    "source": ["A","A","B"],
    "target": ["B","C","C"]
})

G = nx.from_pandas_edgelist(df, "source", "target")

nx.draw(G, with_labels=True)
plt.show()
```
