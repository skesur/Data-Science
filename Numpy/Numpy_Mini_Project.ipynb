```python
import numpy as np
```

```python
students = np.array([
    [85, 78, 92],
    [70, 65, 80],
    [90, 88, 95],
    [60, 70, 75],
    [95, 92, 98]
])

print("Student Marks:\n", students)
```

```python
# Average marks per student
avg_per_student = np.mean(students, axis=1)

# Average marks per subject
avg_per_subject = np.mean(students, axis=0)

print("Average per student:", avg_per_student)
print("Average per subject:", avg_per_subject)
```

```python
topper_index = np.argmax(avg_per_student)
topper_marks = students[topper_index]

print("Topper Index:", topper_index)
print("Topper Marks:", topper_marks)
```

```python
def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    else:
        return "Fail"

grades = np.array([get_grade(avg) for avg in avg_per_student])

print("Grades:", grades)
```

```python
# Students scoring above 80 average
high_performers = students[avg_per_student > 80]

print("High Performers:\n", high_performers)
```

```python
max_per_subject = np.max(students, axis=0)

print("Highest in each subject:", max_per_subject)
```

```python
# Normalize between 0 and 1
normalized = students / 100

print("Normalized Data:\n", normalized)
```

```python
# Normalize between 0 and 1
normalized = students / 100

print("Normalized Data:\n", normalized)
```

```python
# Sort students by average marks
sorted_indices = np.argsort(avg_per_student)[::-1]

print("Ranking (best to worst):", sorted_indices)
```

```python
bonus = np.array([5, 5, 5])

updated_scores = students + bonus

print("Updated Scores:\n", updated_scores)
```

```python
print("\n===== FINAL REPORT =====")
for i in range(len(students)):
    print(f"Student {i+1}: Avg={avg_per_student[i]:.2f}, Grade={grades[i]}")
```
