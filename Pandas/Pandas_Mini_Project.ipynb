```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Product": ["Laptop", "Phone", "Tablet", "Laptop", "Phone", "Tablet"],
    "Price": [50000, 20000, np.nan, 52000, 21000, 15000],
    "Quantity": [2, 5, 3, np.nan, 4, 2],
    "City": ["Ahmedabad", "Delhi", "Mumbai", "Delhi", "Mumbai", "Ahmedabad"]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.isna())
print(df.isna().sum())
```

```python
# Fill Price with mean
df["Price"] = df["Price"].fillna(df["Price"].mean())

# Fill Quantity with median
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())

print(df)
```

```python
df["Revenue"] = df["Price"] * df["Quantity"]

print(df)
```

```python
print("Total Revenue:", np.sum(df["Revenue"]))
print("Average Revenue:", np.mean(df["Revenue"]))
print("Max Revenue:", np.max(df["Revenue"]))
```

```python
city_sales = df.groupby("City")["Revenue"].sum()

print(city_sales)
```

```python
product_sales = df.groupby("Product")["Revenue"].sum()

best_product = product_sales.idxmax()

print("Best Product:", best_product)
```

```python
high_sales = df[df["Revenue"] > 80000]

print(high_sales)
```

```python
df["Rank"] = df["Revenue"].rank(ascending=False)

print(df.sort_values(by="Rank"))
```

```python
revenue_array = df["Revenue"].to_numpy()

print("NumPy Array:", revenue_array)

# Example: normalize
normalized = (revenue_array - np.min(revenue_array)) / (np.max(revenue_array) - np.min(revenue_array))

print("Normalized Revenue:", normalized)
```

```python
print("\n===== SALES REPORT =====")

print("Total Revenue:", df["Revenue"].sum())
print("Best Product:", best_product)
print("Top City:", city_sales.idxmax())

print("\nCity-wise Sales:\n", city_sales)
```
